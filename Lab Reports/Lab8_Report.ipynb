Task: Using Lambda Functions in Electrical Calculations

Objective

The objective of this task is to introduce students to Python lambda functions and demonstrate their use for quick, one-line calculations. Students will create small tools for electrical analysis, such as adjusted voltage metrics and equivalent resistance calculations.

Task Description

The goal of this task is to write lambda functions to perform electrical calculations efficiently. Students will implement simple formulas and test them with given values to understand the usefulness of lambda functions in compact computations.


Task 1 — Basic lambda for a simple electrical formula

Wrote a lambda function that takes a voltage value V and returns V² + 2.

Tested the lambda function with V = 5.

Printed the result to represent a toy “adjusted voltage metric”.



```python
adjust_voltage = lambda V: V**2 + 2
print(adjust_voltage(5))
```

Task 2 — Lambda with two electrical parameters

Wrote a lambda function that takes two resistances R1 and R2 and returns their equivalent parallel resistance.

Tested it with R1 = 100 Ω and R2 = 200 Ω.

Printed the result to verify correctness.

```python
parallel_resistance = lambda R1, R2: (R1 * R2) / (R1 + R2)
print(parallel_resistance(100, 200))
```
