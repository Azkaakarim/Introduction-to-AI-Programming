You are developing a simple energy consumption logging system for a residential building.
Daily electricity usage is stored in a CSV file. so it can be easily read, updated, and organized.


Task 1: Create and Write Energy Data to a CSV File

Problem: You need to create a file and write daily energy consumption data in CSV format.

```python
import csv
file=open ("Energy_Data.csv","w",newline="")
writer=csv.writer(file)
writer.writerow(["Date","Energy(KWh)"])
writer.writerow(["01-12_2025",12])
writer.writerow(["02-12_2025",24])
file.close()
```

Task 2: Read and Display CSV File Data

Problem: Read the contents of energy_data.csv and display all records.

Solution: Use csv.reader to read the file row by row.

```python
import csv
file=open("Energy_Data.csv","r")
reader=csv.reader(file)
for row in reader:
    print(row)

file.close()
```

Task 3: Append New Energy Records

Problem: Add new energy records without deleting existing data.

Solution: Open the file in append mode "a".

```python
import csv
file=open("Energy_Data.csv","a",newline="")
writer= csv.writer(file)
writer.writerow(["03-12-2025",36])
writer.writerow(["04-12-2025",48])
file.close()
print("Row Successfully Added!")
```
