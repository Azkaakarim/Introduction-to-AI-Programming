# This program loads load and price CSVs and calculates hourly and total bills

```python
import csv

hrs = []
load = []
price = []
```

```python
with open("load.csv") as f:
    r = csv.reader(f)
    next(r)
    for row in r:
        hrs.append(int(row[0]))
        load.append(float(row[1]))
```

```python
with open("price.csv") as f:
    r = csv.reader(f)
    next(r)
    for row in r:
        price.append(float(row[1]))
```

```python
def hourly_bill(loads, prices):
    bills = []
    for i in range(len(loads)):
        kwh = loads[i] / 1000
        cost = kwh * prices[i]
        bills.append(cost)
    return bills
```

```python
def total(bills):
    return sum(bills)
```

```python
h_bill = hourly_bill(load, price)
t_bill = total(h_bill)
```

```python
print("Hourly Bill Report ")
for i in range(len(h_bill)):
    print("Hour", hrs[i], ":", h_bill[i], "Rs")

print("\nTotal Bill:", t_bill, "Rs")
```
