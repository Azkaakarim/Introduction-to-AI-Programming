# This program groups consumption by hour and by day from a CSV file
# and calculates averages and totals.

```python

import csv

days = []
hours = []
consumptions = []
```

```python
with open("assignment.csv") as file:
    reader = csv.reader(file)
    next(reader)  
    for row in reader:
        days.append(int(row[0]))
        hours.append(int(row[1]))
        consumptions.append(float(row[2]))
```

```python
hour_groups = {}

for i in range(len(hours)):
    h = hours[i]
    c = consumptions[i]
    if h not in hour_groups:
        hour_groups[h] = []
    hour_groups[h].append(c)
```

```python
print("Consumption Grouped by Hour ")

for h in hour_groups:
    values = hour_groups[h]
    avg = sum(values) / len(values)
    peak = max(values)
    print("Hour:", h)
    print("  Values:", values)
    print("  Average:", avg)
    print("  Peak:", peak)
    print()
```

```python
day_groups = {}
for i in range(len(days)):
    d = days[i]
    c = consumptions[i]
    if d not in day_groups:
        day_groups[d] = []
    day_groups[d].append(c)
```

```python
print("Consumption Grouped by Day ")
for d in day_groups:
    values = day_groups[d]
    total = sum(values)
    avg = total / len(values)
    print("Day:", d)
    print("  Values:", values)
    print("  Total:", total)
    print("  Average:", avg)
    print()
```
