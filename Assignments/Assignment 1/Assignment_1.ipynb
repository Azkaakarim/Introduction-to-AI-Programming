# Program to read energy consumption from CSV and calculate basic values.

```python
import csv

consumption = []
```

```python
with open("assignment.csv") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        consumption.append(float(row[1]))
```

```python
total_consumption = sum(consumption)
average_consumption = total_consumption / len(consumption)
maximum_consumption = max(consumption)
minimum_consumption = min(consumption)

consumption_kwh = [x / 1000 for x in consumption]
```

```python
print("Total Consumption:", total_consumption)
print("Average Consumption:", average_consumption)
print("Maximum Consumption:", maximum_consumption)
print("Minimum Consumption:", minimum_consumption)
print("Consumption in kWh:", consumption_kwh)
```
