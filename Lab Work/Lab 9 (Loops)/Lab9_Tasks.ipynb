For Loop

Task 1 — Print resistor values

Create list of resistors.

Loop through list and print each value.

```python
values=[10,220,56,780.100,340,210]
for v in values:
    print(v)
```

Task 2 — Square current readings

Create list of currents.

Loop and print I² for each current.

```python
currents=[110,78,89,220,120,56,11]
for a in currents:
    power_loss= a*a
    print(power_loss)
```

Task 3 — Generate serial numbers

Use range(1,11).

Loop and print each serial number.

```python
for serial_num in range(1,11):
    print(serial_num)
```

Task 4 — Print voltage values

Create list of voltages.

Loop and print each value.

```python
voltages=[220,11,0,110,560]
for v in voltages:
    print(v)

```

Task 5 — Double capacitor values

Create list of capacitors.

Loop and print double of each value.

```python
capacitances=[10,45,11,34,6,89] #µF
for c in capacitances:
    double=c*2
    print(double)
```

Task 6 — Print sensor temperatures

Create list of temperatures.

Loop and print each sensor’s value.

```python
temps=[23.9,43,50,10.5]
for t in temps:
    print(t)
```

Task 7 — Half solar energy readings

Create list of daily energy.

Loop and print half of each value.

```python
energy_usage=[5600,2400,12,10,6] #kWh
for e in energy_usage:
     half=e/2
     print(half)
```

While Loop

Task 1 — Simulate battery charging

Start battery = 0.

While battery < 100: print "Charging…" and add 10.

```python
battery=0
while battery<100:
    print("charging")
    battery+=10
```

Task 2 — Countdown timer

Start timer = 10.

While timer ≥ 0: print "Test Running" and decrease timer.

```python
timer=10
while timer>=0:
    print("Test Running:",timer)
    timer-=1
```

Task 3 — Voltage input until 0

Use input() repeatedly.

Stop loop when user enters 0.

```python
voltage=1
while voltage!=0:
    voltage=float(input("Enter voltage:"))
    if voltage !=0:
        print("voltage entered was:",voltage)
```

Task 4 — Power doubling simulation

Start power = 1.

While power ≤ 1000: double power each iteration.

```python
power=1
while power <=1000:
    print("power:", power,"w")
    power = power * 2 
print("Power limit exceeded!")
```

Task 5 — Cycle running simulation

Start cycle = 1.

While cycle ≤ 5: print "Cycle Running" and increment cycle.

```python
cycle=1
while cycle <=5:
    print("cycle running:",cycle)
    cycle +=1
print("process completed!")
```

Task 6 — Motor speed simulation

Start speed = 0.

While speed ≤ 3000: add 500 and print each speed.

```python
cycle=0
while cycle<=3000:
    print("Speed",cycle, "RPM")
    cycle+=500
print("Limit reached!")  
```
