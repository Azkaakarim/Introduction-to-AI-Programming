Task 1: Type Error (Add int and string)  

Step 1: Run This Code


```python
a = 10
b = "5"
result = a + b
print(result)
```

Fixed Code:

```python
a = 10
b = "5"
try:
    result = a + b
    print(result)
except TypeError:
    print("Type Error Cannot add an integer and a string")
```

Task 2: Division by Zero  

Step 1: Run This Code


```python
x = 20
y = 0
result = x / y
print(result)
```

Fixed Code:

```python
x = 20
y = 0
try:
    result = x / y
    print(result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed")
```

Task 3: List Index Out of Range  

Step 1: Run This Code


```python
numbers = [10, 20, 30]
print(numbers[5])
```

Fixed Code:

```python
numbers = [10, 20, 30]
try:
    print(numbers[5])
except IndexError:
    print("Error! Index is out of range")
```

Task 4: Logical Error Using raise  

Step 1: Run This Code


```python
age = -5
print("Age:", age)
```

Fixed Code:

```python
age = -5

if age < 0:
    raise Exception("Invalid age value")

print("Age:", age)
```

Task 5: Assert Failure  

Step 1: Run This Code


```python
total_marks = 0
assert total_marks > 0
print("Total marks:", total_marks)
```

Fixed Code:

```python
total_marks = 0

try:
    assert total_marks > 0
    print("Total marks:", total_marks)
except AssertionError:
    print("Assertion Error: Total marks must be greater than zero")
```
