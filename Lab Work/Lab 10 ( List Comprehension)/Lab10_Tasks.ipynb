Task 1: Filter Voltage Readings (< 200 V)  


Takes the list of voltages

Selects only values below 200 volts

Stores the filtered values in a new list

```python
voltages = [224, 198, 231, 177, 205, 189, 240]
low_voltages = [v for v in voltages if v < 200]
print(low_voltages)
```

Task 2: Convert Power Ratings to kW  

Takes power values in watts

Converts each value to kilowatts

Creates a new list of converted values

```python
powers_w = [50, 120, 300, 750, 1500]
powers_kw = [p / 1000 for p in powers_w]
print(powers_kw)
```

Task 3: Square Only Even Resistances  


Checks each resistor value

Selects only even numbers

Squares them and saves the results

```python
resistors = [10, 15, 22, 47, 50, 68]
even_squares = [r*r for r in resistors if r % 2 == 0]
print(even_squares)
```

Task 4: Label Currents as “High” or “Low”  


Reads each current value

Compares it with 1.5 A

Assigns “High” if above 1.5, otherwise “Low”

```python
currents = [2.1, 0.5, 1.8, 3.0, 0.9]
labels = ["High" if c > 1.5 else "Low" for c in currents]
print(labels)
```

Task 5: Extract Appliance Names Starting With ‘F’  


Reads appliance names

Selects names beginning with the letter F

```python
appliances = ["Fan", "Fridge", "Light", "Freezer", "Pump"]
f_appliances = [a for a in appliances if a.startswith("F")]
print(f_appliances)
```

Task 6: Generate First 10 Multiples of 50  


Generates values from 1 to 10

Multiplies each by 50

Produces multiples from 50 to 500

```python
multiples_50 = [50*i for i in range(1, 11)]
print(multiples_50)
```
